```python
try:
    model = CNN_double()
except Exception as e:
    print(f'Error initializing CNN_double model: {e}')
try:
    dir=os.getcwd()
except Exception as e:
    print(f'Error getting current working directory: {e}')
try:
    train_gen, val_gen = get_pcam_generators(dir+'\\Data')
except Exception as e:
    print(f'Error getting PCAM generators: {e}')
try:
    model_name='double_CNN_50epochs'
except Exception as e:
    print(f'Error setting model name: {e}')
try:
    model , callbacks_list, model_path = model_preparation(model,model_name,dir)
    if os.path.exists(model_path):
        print("Checkpoint found. Loading weights...")
        model.load_weights(model_path)
    else:
        print("No checkpoint found. Starting training from scratch.")
except Exception as e:
    print(f'Error preparing model: {e}')
try:
    epochs=50
except Exception as e:
    print(f'Error setting epochs: {e}')
try:
    model, history = train_model(model, train_gen, val_gen, epochs, callbacks_list)
except Exception as e:
    print(f'Error training model: {e}')
try:
    loss, accuracy, area_under_curve, recall = evaluate_model_from_file(model_path, val_gen)
except Exception as e:
    print(f'Error evaluating model: {e}')
```
