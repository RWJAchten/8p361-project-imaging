plot validation accuracy over epochs for all models

```python
import os
import json
import matplotlib.pyplot as plt
import pandas as pd

# Function to load a JSON file and convert it into a DataFrame
def load_json_to_dataframe(file_path):
    with open(file_path, "r") as file:
        data = json.load(file)  # Read the JSON file (list of lists)

    # Check if data is in expected format: list of lists with three elements (timestamp, index, accuracy)
    if isinstance(data, list) and all(isinstance(row, list) and len(row) == 3 for row in data):
        df = pd.DataFrame(data, columns=["timestamp", "index", "accuracy"])
    else:
        raise ValueError(f"Unexpected JSON structure in file: {file_path}")

    return df

# Function to read multiple JSON files and plot index (x-axis) vs accuracy (y-axis)
def plot_multiple_json(file_paths):
    plt.figure(figsize=(10, 6))  # Create a larger figure

    colors = {}  # Dictionary to store colors per model group
    
    for file_path in file_paths:
        df = load_json_to_dataframe(file_path)

        # Extract filename without path and remove ".json"
        label = os.path.basename(file_path).replace(".json", "")

        # Determine whether the file is for training or validation
        is_training = "train" in label
        base_label = label.replace("_train", "").replace("_validation", "")

        # Assign a unique color for each model group
        if base_label not in colors:
            colors[base_label] = plt.cm.tab10(len(colors) % 10)

        # Choose linestyle based on train/validation
        linestyle = "--" if is_training else "-"

        # Plot (index on x-axis, accuracy on y-axis)
        plt.plot(df["index"], df["accuracy"], marker="", linestyle=linestyle, color=colors[base_label], label=label)

    # Set graph properties
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Accuracy for all models")
    plt.legend()
    plt.xlim(0, 49)  # Limit x-axis from 0 to 49 epochs
    plt.grid(True)
    plt.show()

# Define the list of file paths
dir_path = os.path.join(os.getcwd(), "ACCURACY")
file_paths = [
    os.path.join(dir_path, "double_CNN_50epochs_train.json"), os.path.join(dir_path, "double_hybrid_50epochs_train.json"),
    os.path.join(dir_path, "double_VIT_50epochs_2_train.json"), os.path.join(dir_path, "single_hybrid_50epochs_train.json"),
    os.path.join(dir_path, "double_CNN_50epochs_validation.json"), os.path.join(dir_path, "double_hybrid_50epochs_validation.json"),
    os.path.join(dir_path, "double_VIT_50epochs_2_validation.json"), os.path.join(dir_path, "single_hybrid_50epochs_validation.json")
]

# Call the function to plot data
plot_multiple_json(file_paths)



```
